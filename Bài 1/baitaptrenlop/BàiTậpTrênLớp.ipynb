```python

 #-*- coding: utf-8 -*-
"""
Bài Tập Trên Lớp

@author: Administrator
"""
import re
#bài tập 12
def tron_mang(a, b):
    s = []
    n = min(len(a), len(b))

    for i in range(n):
        s.append(a[i] + b[i])

    # Thêm phần dư
    if len(a) > n:
        s.extend(a[n:])
    if len(b) > n:
        s.extend(b[n:])

    return s

a = [36, 3, 6, 18]
b = [2, 7, 4, 3, 2, 8]

print(tron_mang(a, b))
```

```python
import random

# a. Tạo mảng
def tao_mang(m, n):
    return [[random.randint(1, 100) for _ in range(n)] for _ in range(m)]

# Xuất mảng
def xuat_mang(a):
    for row in a:
        print(row)

# b. Xuất dòng k
def xuat_dong(a, k):
    print(a[k])

# c. Xuất cột k
def xuat_cot(a, k):
    for row in a:
        print(row[k])

# d. Dòng có tổng lớn nhất
def dong_max(a):
    return max(range(len(a)), key=lambda i: sum(a[i]))

# e. Cột có tích nhỏ nhất
def cot_min_tich(a):
    n = len(a[0])
    min_col = 0
    min_product = float('inf')

    for j in range(n):
        product = 1
        for i in range(len(a)):
            product *= a[i][j]
        if product < min_product:
            min_product = product
            min_col = j

    return min_col

# f. Dòng chẵn, cột lẻ
def chan_le(a):
    for i in range(len(a)):
        for j in range(len(a[0])):
            if i % 2 == 0 and j % 2 == 1:
                print(a[i][j], end=" ")
    print()

# g. TBC phần tử chẵn dòng lẻ
def tbc_chan_dong_le(a):
    s = 0
    count = 0
    for i in range(len(a)):
        if i % 2 == 1:
            for x in a[i]:
                if x % 2 == 0:
                    s += x
                    count += 1
    return s / count if count else 0

# h. TBC biên
def tbc_bien(a):
    s = 0
    count = 0
    m, n = len(a), len(a[0])

    for i in range(m):
        for j in range(n):
            if i == 0 or j == 0 or i == m-1 or j == n-1:
                s += a[i][j]
                count += 1

    return s / count

# i. TBC không biên
def tbc_khong_bien(a):
    s = 0
    count = 0
    m, n = len(a), len(a[0])

    for i in range(1, m-1):
        for j in range(1, n-1):
            s += a[i][j]
            count += 1

    return s / count if count else 0


# ===== MENU =====
a = []

while True:
    print("\n===== MENU BÀI 14 =====")
    print("1. Tạo mảng")
    print("2. Xuất mảng")
    print("3. Xuất dòng k")
    print("4. Xuất cột k")
    print("5. Tìm dòng có tổng lớn nhất")
    print("6. Tìm cột có tích nhỏ nhất")
    print("7. Xuất phần tử dòng chẵn, cột lẻ")
    print("8. TBC phần tử chẵn dòng lẻ")
    print("9. TBC các phần tử biên")
    print("10. TBC phần tử không biên")
    print("0. Thoát")

    chon = input("Nhập lựa chọn: ")

    if chon == "1":
        m = int(input("Nhập số dòng: "))
        n = int(input("Nhập số cột: "))
        a = tao_mang(m, n)
        print("Đã tạo mảng!")

    elif chon == "2":
        if a:
            xuat_mang(a)
        else:
            print("Chưa có mảng!")

    elif chon == "3":
        k = int(input("Nhập dòng k: "))
        xuat_dong(a, k)

    elif chon == "4":
        k = int(input("Nhập cột k: "))
        xuat_cot(a, k)

    elif chon == "5":
        print("Dòng có tổng lớn nhất là:", dong_max(a))

    elif chon == "6":
        print("Cột có tích nhỏ nhất là:", cot_min_tich(a))

    elif chon == "7":
        chan_le(a)

    elif chon == "8":
        print("TBC:", tbc_chan_dong_le(a))

    elif chon == "9":
        print("TBC biên:", tbc_bien(a))

    elif chon == "10":
        print("TBC không biên:", tbc_khong_bien(a))

    elif chon == "0":
        print("Thoát!")
        break

    else:
        print("Lựa chọn không hợp lệ!")
```

```python
import re
import datetime

# bai 15
class SinhVien:
    def __init__(self, masv, tensv, diem, namsinh):
        self.masv = masv
        self.tensv = tensv
        self.diem = diem
        self.namsinh = namsinh

    def xuat(self):
        print("Danh Sách Sinh Viên")
        print(f"Mã Sinh Viên: {self.masv}\tTên SV : {self.tensv}\tNăm sinh: {self.namsinh}\tĐTB : {self.diem:.2f}")

# Functions outside the class
def nhap_danh_sach(n):
    ds = []
    for i in range(n):
        print(f"\nNhập Danh Sách Thứ {i+1}:")
        maSV = input("Nhập Mã Sinh Viên: ")
        tenSV = input("Nhập Tên Sinh Viên: ")
        namSinh = int(input("Nhập Năm Sinh: "))
        DiemTB = float(input("Nhập Điểm Trung Bình: ").replace(',', '.'))
        ds.append(SinhVien(maSV, tenSV, DiemTB, namSinh))
    return ds

def demlenlop(ds):
    return sum(1 for sv in ds if sv.diem >= 5)

def ktratuoisv(ds):
    year = datetime.datetime.now().year
    found_student = False
    for sv in ds:
        if year - sv.namsinh >= 20:
            sv.xuat()
            found_student = True
    if not found_student:
        print("Không Có Sinh Viên Nào Lớn Hơn 20 tuổi")

def xeploai(ds, min_diem):
    found_student = False
    for sv in ds:
        if sv.diem >= min_diem:
            sv.xuat()
            found_student = True
    if not found_student:
        print("Không Có Sinh Viên Nào đạt điểm yêu cầu")

def ktramsv(ds):
    dem = 0
    for sv in ds:
        # Assuming 'DH' at the beginning of the student ID
        if len(sv.masv) >= 2 and sv.masv.startswith("DH"):
            dem += 1
    return dem

def dem_ten_Lan(ds):
    return sum(1 for sv in ds if "Lan" in sv.tensv)

def dem_ho_Phan(ds):
    return sum(1 for sv in ds if sv.tensv.startswith("Phan")) # Assuming 'Phan' is a first name/surname

def xuat_ds(ds):
    if not ds:
        print("Danh sách sinh viên rỗng.")
        return
    for sv in ds:
        sv.xuat()

def menu():
    n = int(input("Nhap so luong SV: "))
    ds = nhap_danh_sach(n)

    while True:
        print("\n===== MENU =====")
        print("1. Xuat danh sach SV")
        print("2. Dem SV len lop")
        print("3. Xuat SV 20 tuoi tro len")
        print("4. Dem SV he DH")
        print("5. Dem SV ten Lan")
        print("6. Dem SV ho Phan")
        print("7. Xep loai sinh vien theo diem")
        print("0. Thoat")

        chon = input("Chon: ") # Read as string to avoid ValueError on non-int input

        if chon == "1":
            xuat_ds(ds)
        elif chon == "2":
            print("So SV len lop:", demlenlop(ds))
        elif chon == "3":
            print("SV du 20 tuoi tro len:")
            ktratuoisv(ds)
        elif chon == "4":
            print("So SV he DH:", ktramsv(ds))
        elif chon == "5":
            print("So SV ten Lan:", dem_ten_Lan(ds))
        elif chon == "6":
            print("So SV ho Phan:", dem_ho_Phan(ds))
        elif chon == "7":
            min_diem = float(input("Nhap diem toi thieu de xep loai: "))
            xeploai(ds, min_diem)
        elif chon == "0":
            print("Thoat!")
            break
        else:
            print("Chon sai!")

menu()
```
