```python
import copy
from heapq import heappush, heappop

n = 3
dong = [1, 0, -1, 0]
cot = [0, -1, 0, 1]

class hangdoi:
    def __init__(self):
        self.heap = []

    def push(self, tkhoa):
        heappush(self.heap, tkhoa)

    def pop(self):
        return heappop(self.heap)

    def ktra(self):
        return not self.heap


class nodes:
    def __init__(self, nutdau, matran, vitri_o_trong, giatri, lop):
        self.nutdau = nutdau
        self.matran = matran
        self.vitri_o_trong = vitri_o_trong
        self.giatri = giatri
        self.lop = lop

    # sửa __It__ -> __lt__
    def __lt__(self, nxt):
        return (self.giatri + self.lop) < (nxt.giatri + nxt.lop)


# đưa ra ngoài class để gọi dễ
def heuristic(matran, vitri_cuoicung):
    dem = 0
    for i in range(n):
        for j in range(n):
            if matran[i][j] != 0 and matran[i][j] != vitri_cuoicung[i][j]:
                dem += 1
    return dem


def newnodes(matran, vitri_o_trong, otrong_moi, lop, nutcha, vitri_cuoicung):
    matranmoi = copy.deepcopy(matran)

    a1, b1 = vitri_o_trong
    a2, b2 = otrong_moi

    # hoán đổi
    matranmoi[a1][b1], matranmoi[a2][b2] = matranmoi[a2][b2], matranmoi[a1][b1]

    giatri = heuristic(matranmoi, vitri_cuoicung)

    node_moi = nodes(nutcha, matranmoi, otrong_moi, giatri, lop)
    return node_moi


def inmatran(matran):
    for i in range(n):
        for j in range(n):
            print("%d" % matran[i][j], end=" ")
        print()
    print()


def luugiatri(x, y):
    return x >= 0 and x < n and y >= 0 and y < n


def printtrang(root):
    if root is None:
        return
    printtrang(root.nutdau)
    inmatran(root.matran)


def giaiphap(initial, vitri_o_trong, vitri_cuoicung):
    pq = hangdoi()

    giatri = heuristic(initial, vitri_cuoicung)
    root = nodes(None, initial, vitri_o_trong, giatri, 0)

    pq.push(root)

    while not pq.ktra():
        minimum = pq.pop()

        if minimum.giatri == 0:
            print(" Solution:\n")
            printtrang(minimum)
            return

        for i in range(4):
            vitriomoi = (
                minimum.vitri_o_trong[0] + dong[i],
                minimum.vitri_o_trong[1] + cot[i]
            )

            if luugiatri(vitriomoi[0], vitriomoi[1]):
                nodecon = newnodes(
                    minimum.matran,
                    minimum.vitri_o_trong,
                    vitriomoi,
                    minimum.lop + 1,
                    minimum,
                    vitri_cuoicung
                )
                pq.push(nodecon)


# TEST
initial = [
    [1, 2, 3],
    [5, 6, 0],
    [7, 8, 4]
]

vitri_cuoicung = [
    [1, 2, 3],
    [4, 5, 6],
    [0, 8, 7]
]

vitri_o_trong = (1, 2)
giaiphap(initial, vitri_o_trong, vitri_cuoicung)
```
