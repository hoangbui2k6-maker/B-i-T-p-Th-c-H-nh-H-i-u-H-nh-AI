```python
#bai1
import math

def tao_ban_co():
    return [[" " for _ in range(3)] for _ in range(3)]

def kiem_tra_thang(ban_co, nguoi):
    for i in range(3):
        if all(ban_co[i][j] == nguoi for j in range(3)):
            return True
        if all(ban_co[j][i] == nguoi for j in range(3)):
            return True
    if all(ban_co[i][i] == nguoi for i in range(3)):
        return True
    if all(ban_co[i][2 - i] == nguoi for i in range(3)):
        return True
    return False

def la_hoa(ban_co):
    return all(ban_co[i][j] != " " for i in range(3) for j in range(3))

def minimax(ban_co, do_sau, la_max):
    if kiem_tra_thang(ban_co, "X"):
        return 1
    if kiem_tra_thang(ban_co, "O"):
        return -1
    if la_hoa(ban_co):
        return 0

    if la_max:
        best = -math.inf
        for i in range(3):
            for j in range(3):
                if ban_co[i][j] == " ":
                    ban_co[i][j] = "X"
                    score = minimax(ban_co, do_sau + 1, False)
                    ban_co[i][j] = " "
                    best = max(best, score)
        return best
    else:
        best = math.inf
        for i in range(3):
            for j in range(3):
                if ban_co[i][j] == " ":
                    ban_co[i][j] = "O"
                    score = minimax(ban_co, do_sau + 1, True)
                    ban_co[i][j] = " "
                    best = min(best, score)
        return best

def nuoc_di_tot_nhat(ban_co):
    best_val = -math.inf
    move = (-1, -1)
    for i in range(3):
        for j in range(3):
            if ban_co[i][j] == " ":
                ban_co[i][j] = "X"
                move_val = minimax(ban_co, 0, False)
                ban_co[i][j] = " "
                if move_val > best_val:
                    move = (i, j)
                    best_val = move_val
    return move
```

```python
#bai 2
def alpha_beta(ban_co, do_sau, alpha, beta, la_max):
    if kiem_tra_thang(ban_co, "X"):
        return 1
    if kiem_tra_thang(ban_co, "O"):
        return -1
    if la_hoa(ban_co):
        return 0

    if la_max:
        best = -math.inf
        for i in range(3):
            for j in range(3):
                if ban_co[i][j] == " ":
                    ban_co[i][j] = "X"
                    val = alpha_beta(ban_co, do_sau + 1, alpha, beta, False)
                    ban_co[i][j] = " "
                    best = max(best, val)
                    alpha = max(alpha, best)
                    if beta <= alpha:
                        break
        return best
    else:
        best = math.inf
        for i in range(3):
            for j in range(3):
                if ban_co[i][j] == " ":
                    ban_co[i][j] = "O"
                    val = alpha_beta(ban_co, do_sau + 1, alpha, beta, True)
                    ban_co[i][j] = " "
                    best = min(best, val)
                    beta = min(beta, best)
                    if beta <= alpha:
                        break
        return best

def nuoc_di_tot_nhat_ab(ban_co):
    best_val = -math.inf
    move = (-1, -1)
    for i in range(3):
        for j in range(3):
            if ban_co[i][j] == " ":
                ban_co[i][j] = "X"
                move_val = alpha_beta(ban_co, 0, -math.inf, math.inf, False)
                ban_co[i][j] = " "
                if move_val > best_val:
                    move = (i, j)
                    best_val = move_val
    return move
```
